# 3. Label Encoding for Categorical Variables

```python
#Convert categorical   such as Category, Sub Category, City, Region,
#State, and Month into numerical values.
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

# encode categorical variables
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df=pd.read_csv("Supermart Grocery Sales.csv")
df.columns
```

```python
# Initialize the label encoder
le = LabelEncoder()


df['Category']=le.fit_transform(df['Category'])
df['Sub Category']=le.fit_transform(df['Sub Category'])
df['City']=le.fit_transform(df['City'])
df['Region']=le.fit_transform(df['Region'])
df['State']=le.fit_transform(df['State'])
```

# Step 5: Feature Selection and Model Building

```python
# select features and target columns
df.columns
```

```python
# in the above columns we use feature matrix variable'Category', 'Sub Category', 'City','Region''Discount', 'Profit',
# and reaming columns we will drop
feature_data=df.drop(columns=['Order ID','Customer Name','Order Date','Sales'])
```

```python
target_data=df['Sales']
```

```python
# now split the data in train and test 
x_train,x_test,y_train,y_test=train_test_split(feature_data,target_data,test_size=0.2,random_state=40)
# feautre_sacling
scaler = StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

# Step 6: Train a Linear Regression Model

```python
# Initialize the model
model=LinearRegression()
# train the model
model.fit(x_train,y_train)
# make prediction
pred=model.predict(x_test)
```

# Step 7: Evaluate the Model

```python
# Evaluate the model performance using Mean Squared Error (MSE) and R-squared
# calculate mean square error and r-squared
mse = mean_squared_error(y_test, pred)
r2 = r2_score(y_test, pred)
print(f"mean_squared_error:{mse}")
print(f"r_squared:{r2}")
```

# Step 8: Visualize the Results


```python
# Actual vs Predicted Sales
plt.figure(figsize=(8, 6))
plt.scatter(y_test, pred)
plt.plot([min(y_test), max(y_test)], [min(y_test),max(y_test)], color='red')
plt.title('Actual vs Predicted Sales')
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.show()
```
